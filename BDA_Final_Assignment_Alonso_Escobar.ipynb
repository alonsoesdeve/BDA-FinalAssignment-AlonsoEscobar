```python
import pandas as pd
import numpy as np

df2 = df.copy()

na_like = {"", " ", "nan", "NaN", "NULL", "null", "None", "none", "Na"}
df2 = df2.applymap(lambda x: np.nan if isinstance(x, str) and x.strip() in na_like else x)

for c in df2.select_dtypes(include="object").columns:
    df2[c] = df2[c].astype(str).str.strip()

numeric_targets = [
    "Delivery_person_Age", "Delivery_person_Ratings",
    "Restaurant_latitude", "Restaurant_longitude",
    "Delivery_location_latitude", "Delivery_location_longitude",
    "Vehicle_condition", "multiple_deliveries"
]
for c in numeric_targets:
    if c in df2.columns:
        df2[c] = pd.to_numeric(df2[c], errors="coerce")

if "Order_Date" in df2:
    df2["Order_Date"] = pd.to_datetime(df2["Order_Date"], errors="coerce", dayfirst=True)

def parse_time_series(series: pd.Series) -> pd.Series:
    s = series.astype(str)
    t = pd.to_datetime(s, format="%H:%M:%S", errors="coerce")
    miss = t.isna()
    if miss.any():
        t.loc[miss] = pd.to_datetime(s[miss], format="%H:%M", errors="coerce")
    return t.dt.time

for c in ["Time_Orderd", "Time_Order_picked"]:
    if c in df2.columns:
        df2[c] = parse_time_series(df2[c])

for c in ["Time_Orderd", "Time_Order_picked"]:
    dtc = c + "_dt"
    if {"Order_Date", c}.issubset(df2.columns):
        df2[dtc] = pd.to_datetime(
            df2["Order_Date"].astype(str) + " " + df2[c].astype(str),
            errors="coerce"
        )
```

```python
PATH = "test.csv"

df = (spark.read.option("header", True).csv(PATH))


for c, t in df.dtypes:
    if t == "string":
        df = df.withColumn(c, F.trim(F.col(c)))


num_cols = ["Delivery_person_Age","Delivery_person_Ratings",
            "Restaurant_latitude","Restaurant_longitude",
            "Delivery_location_latitude","Delivery_location_longitude",
            "Vehicle_condition","multiple_deliveries"]
for c in num_cols:
    if c in df.columns:
        df = df.withColumn(c, F.col(c).cast("double"))


if "Order_Date" in df.columns:
    df = df.withColumn("Order_Date", F.to_date("Order_Date", "dd-MM-yyyy"))

def to_ts(date_col, time_col):
    return F.coalesce(
        F.to_timestamp(F.concat_ws(" ", F.date_format(date_col, "dd-MM-yyyy"), time_col), "dd-MM-yyyy HH:mm:ss"),
        F.to_timestamp(F.concat_ws(" ", F.date_format(date_col, "dd-MM-yyyy"), time_col), "dd-MM-yyyy HH:mm")
    )

if set(["Order_Date","Time_Orderd"]).issubset(df.columns):
    df = df.withColumn("Time_Orderd_dt", to_ts(F.col("Order_Date"), F.col("Time_Orderd")))
if set(["Order_Date","Time_Order_picked"]).issubset(df.columns):
    df = df.withColumn("Time_Order_picked_dt", to_ts(F.col("Order_Date"), F.col("Time_Order_picked")))


df = df.withColumn("prep_minutes",
    (F.col("Time_Order_picked_dt").cast("long") - F.col("Time_Orderd_dt").cast("long"))/60.0
)


def haversine_expr(lat1, lon1, lat2, lon2):
    r = 6371.0
    return (F.lit(2*r) * F.asin(F.sqrt(
        F.pow(F.sin((F.radians(lat2) - F.radians(lat1))/2), 2) +
        F.cos(F.radians(lat1)) * F.cos(F.radians(lat2)) *
        F.pow(F.sin((F.radians(lon2) - F.radians(lon1))/2), 2)
    )))
df = df.withColumn("km_rest_to_drop", haversine_expr(
    F.col("Restaurant_latitude"), F.col("Restaurant_longitude"),
    F.col("Delivery_location_latitude"), F.col("Delivery_location_longitude"))
)


df = (df
    .withColumn("order_hour", F.hour("Time_Orderd_dt"))
    .withColumn("order_dow", F.date_format("Order_Date", "EEEE"))
    .withColumn("is_weekend", F.col("order_dow").isin("Saturday","Sunday").cast("int"))
)


def norm_title(col):
    return F.initcap(F.regexp_replace(F.regexp_replace(F.trim(col), r"\s+", " "), "-", " "))
for c in ["Weatherconditions","Road_traffic_density","Type_of_order","Type_of_vehicle","Festival","City"]:
    if c in df.columns:
        df = df.withColumn(c, norm_title(F.col(c)))


if "Festival" in df.columns:
    df = df.withColumn("Festival_flag", F.col("Festival").isin("Yes","True","Festival","Holiday").cast("int"))
if "Road_traffic_density" in df.columns:
    df = df.withColumn("traffic_level",
        F.when(F.col("Road_traffic_density")=="Low",0)
         .when(F.col("Road_traffic_density")=="Medium",1)
         .when(F.col("Road_traffic_density")=="High",2)
         .when(F.col("Road_traffic_density")=="Jam",3)
         .otherwise(None).cast("double")
    )

df = (df
    .withColumn("prep_minutes", F.when((F.col("prep_minutes") < 0) | (F.col("prep_minutes") > 240), None).otherwise(F.col("prep_minutes")))
    .withColumn("km_rest_to_drop", F.when((F.col("km_rest_to_drop") <= 0) | (F.col("km_rest_to_drop") > 100), None).otherwise(F.col("km_rest_to_drop")))
    .withColumn("order_hour", F.when(~F.col("order_hour").between(0,23), None).otherwise(F.col("order_hour")))
)

df.select("prep_minutes","km_rest_to_drop","order_hour","traffic_level","Festival_flag").summary().show()
```

```python
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.ml.regression import LinearRegression, RandomForestRegressor
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml.evaluation import RegressionEvaluator


label_col = "prep_minutes"
base_feats = [
    "km_rest_to_drop","order_hour","is_weekend","Festival_flag","traffic_level",
    "Delivery_person_Age","Delivery_person_Ratings","Vehicle_condition","multiple_deliveries"
]
cat_feats = [c for c in ["City","Weatherconditions","Road_traffic_density","Type_of_order","Type_of_vehicle"] if c in df.columns]
all_feats = [c for c in base_feats if c in df.columns] + cat_feats


df_model = df.dropna(subset=[label_col])


indexers = [StringIndexer(inputCol=c, outputCol=f"{c}_idx", handleInvalid="keep") for c in cat_feats]
encoders = [OneHotEncoder(inputCols=[f"{c}_idx"], outputCols=[f"{c}_oh"], dropLast=True) for c in cat_feats]

assembler_inputs = [c for c in base_feats if c in df_model.columns] + [f"{c}_oh" for c in cat_feats]
assembler = VectorAssembler(inputCols=assembler_inputs, outputCol="features", handleInvalid="skip")


lr = LinearRegression(featuresCol="features", labelCol=label_col, predictionCol="prediction")
rf = RandomForestRegressor(featuresCol="features", labelCol=label_col, predictionCol="prediction", seed=42)


pipe_lr = Pipeline(stages=indexers + encoders + [assembler, lr])
pipe_rf = Pipeline(stages=indexers + encoders + [assembler, rf])


train, test = df_model.randomSplit([0.8, 0.2], seed=42)


grid_lr = (ParamGridBuilder()
           .addGrid(lr.regParam, [0.0, 0.01, 0.1])
           .addGrid(lr.elasticNetParam, [0.0, 0.5, 1.0])
           .build())

grid_rf = (ParamGridBuilder()
           .addGrid(rf.numTrees, [50, 150])
           .addGrid(rf.maxDepth, [6, 10, 14])
           .build())

evaluator = RegressionEvaluator(labelCol=label_col, predictionCol="prediction", metricName="rmse")

cv_lr = CrossValidator(estimator=pipe_lr, estimatorParamMaps=grid_lr, evaluator=evaluator, numFolds=3, seed=42)
cv_rf = CrossValidator(estimator=pipe_rf, estimatorParamMaps=grid_rf, evaluator=evaluator, numFolds=3, seed=42)

cv_lr_model = cv_lr.fit(train)
cv_rf_model = cv_rf.fit(train)


rmse_lr = cv_lr_model.avgMetrics[cv_lr_model.avgMetrics.index(min(cv_lr_model.avgMetrics))]
rmse_rf = cv_rf_model.avgMetrics[cv_rf_model.avgMetrics.index(min(cv_rf_model.avgMetrics))]
best_model = cv_rf_model if min(cv_rf_model.avgMetrics) <= min(cv_lr_model.avgMetrics) else cv_lr_model
print("Best model:", "RandomForest" if best_model is cv_rf_model else "LinearRegression")
print("CV RMSE â€” LR:", min(cv_lr_model.avgMetrics), "| RF:", min(cv_rf_model.avgMetrics))
```

```python
best_pm = best_model.bestModel
pred = best_pm.transform(test).select(label_col, "prediction")
rmse = evaluator.evaluate(pred, {evaluator.metricName: "rmse"})
mae  = RegressionEvaluator(labelCol=label_col, predictionCol="prediction", metricName="mae").evaluate(pred)
r2   = RegressionEvaluator(labelCol=label_col, predictionCol="prediction", metricName="r2").evaluate(pred)
print(f"Test RMSE={rmse:.3f} min | MAE={mae:.3f} min | R^2={r2:.3f}")


out_dir = "/content/ml_outputs"; import os; os.makedirs(out_dir, exist_ok=True)
pred.limit(20).toPandas().to_csv(f"{out_dir}/sample_predictions.csv", index=False)
```

```python
from pyspark.ml.regression import RandomForestRegressionModel, LinearRegressionModel

def get_feature_names_from_metadata(df_with_features, features_col="features"):
    meta = df_with_features.schema[features_col].metadata
    names = []
    if "ml_attr" in meta:
        attrs = meta["ml_attr"]
        for k in ["numeric","binary"]:
            if k in attrs:
                names += [a["name"] for a in attrs[k]]
    return names


sample_feats = best_pm.transform(train.limit(1000)).select("features")
feat_names = get_feature_names_from_metadata(sample_feats)

if isinstance(best_pm.stages[-1], RandomForestRegressionModel):
    rf_model = best_pm.stages[-1]
    imps = rf_model.featureImportances.toArray()

    pairs = list(zip(feat_names if feat_names else range(len(imps)), imps))
    top = sorted(pairs, key=lambda x: x[1], reverse=True)[:20]
    print("\nTop feature importances (RF):")
    for n, w in top:
        print(f"- {n}: {w:.4f}")
elif isinstance(best_pm.stages[-1], LinearRegressionModel):
    lr_model = best_pm.stages[-1]
    coeffs = lr_model.coefficients.toArray()
    pairs = list(zip(feat_names if feat_names else range(len(coeffs)), coeffs))
    top = sorted(pairs, key=lambda x: abs(x[1]), reverse=True)[:20]
    print("\nTop coefficients (LR):")
    for n, w in top:
        print(f"- {n}: {w:.4f}")
```
